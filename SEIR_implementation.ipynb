```python
from SEIR_model import *
```

```python
SIGMA = 1/5.2
GAMMA = 1/10
MU_I  = 0.002

R0    = 6
BETA  = GAMMA * R0 
```

```python
model = SEIRSModel(initN   = 20000000,
                   beta    = BETA, 
                   sigma   = SIGMA, 
                   gamma   = GAMMA, 
                   mu_I    = MU_I,
                   mu_0    = 0, 
                   nu      = 0, 
                   xi      = 0,
                   beta_Q  = 0.5*(BETA), 
                   sigma_Q = SIGMA, 
                   gamma_Q = GAMMA, 
                   mu_Q    = MU_I,
                   theta_E = 0, 
                   theta_I = 0, 
                   psi_E   = 1.0, 
                   psi_I   = 1.0,
                   initI   = 10000, 
                   initE   = 0, 
                   initQ_E = 0, 
                   initQ_I = 0, 
                   initR   = 0, 
                   initF   = 0)
```

```python
checkpoints = {'t':       [15, 90], 
               'beta':    [BETA*0.5, BETA * 0.8], 
               'theta_E': [0.06, 0.04], 
               'theta_I': [0.06, 0.04]
              }
```

```python
model.run(T=300,checkpoints = checkpoints)
ref_model = SEIRSModel(initN   = 20000000,
                   beta    = BETA, 
                   sigma   = SIGMA, 
                   gamma   = GAMMA, 
                   mu_I    = MU_I,
                   mu_0    = 0, 
                   nu      = 0, 
                   xi      = 0,
                   beta_Q  = 0.5*(BETA), 
                   sigma_Q = SIGMA, 
                   gamma_Q = GAMMA, 
                   mu_Q    = MU_I,
                   theta_E = 0, 
                   theta_I = 0, 
                   psi_E   = 1.0, 
                   psi_I   = 1.0,
                   initI   = 10000, 
                   initE   = 0, 
                   initQ_E = 0, 
                   initQ_I = 0, 
                   initR   = 0, 
                   initF   = 0)
ref_model.run(T=300)
```

```python
model.figure_infections(ylim = 0.8,shaded_reference_results= ref_model)
```

```python
model.figure_basic(ylim = 1,shaded_reference_results= ref_model)
```
